```python
import re

def clean_name(df, name_column='Name'):
    titles = r'(Mr\.|Ms\.|Mrs\.|Dr\.|Prof\.|Rev\.|Sir|Madam|Lord|Lady)\s+' # Regex for common titles
    df[name_column] = df[name_column].str.replace(titles, '', regex=True, flags=re.IGNORECASE)
    df[name_column] = df[name_column].str.lower().str.strip() # lowercase and remove leading/trailing spaces
    return df

# Example Usage:
data = {'name': ['Mr. John Smith', 'Ms. Jane Doe', 'PROF. DAVID LEE', '  Lady  Sarah Connor  ']}
df = pd.DataFrame(data)
df.head()
```

```python
df['Name_Cleaned']= clean_name(df)
```
