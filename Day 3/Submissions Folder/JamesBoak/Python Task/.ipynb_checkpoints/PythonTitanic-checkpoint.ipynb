```python
from sklearn import datasets
import seaborn as sns
import pandas as pd
import numpy as np

data = sns.load_dataset("titanic")
df = pd.DataFrame(data)
```

```python
df.head()
df.count()
```

```python
#identify blanks
df.isnull().sum()
```

```python
#remove blanks
df = df.dropna()
df.duplicated().sum()
df = df.drop_duplicates()
df.head()
```

```python
#check sex values
df["sex"].unique()
```

```python
##Handle outliers
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.boxplot(x=df["age"])
plt.title("Age Column")

plt.subplot(1, 2, 2)
sns.boxplot(x=df["fare"])
plt.title("Fare Column")

plt.show()
```

```python
#remove outliers
def remove_outliers_iqr(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

df = remove_outliers_iqr("age")
df = remove_outliers_iqr("fare")

df.count()
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.boxplot(x=df["age"])
plt.title("Age Column After Handling Outliers")

plt.subplot(1, 2, 2)
sns.boxplot(x=df["fare"])
plt.title("Fare Column After Handling Outliers")

plt.show()
```

```python
#add family size column
df["Family_Size"] = df["sibsp"] + df["parch"] + 1
print(df[["sibsp", "parch", "Family_Size"]].head())
```

```python
#encode embarked column
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform the 'Embarked' column
df["embarked"] = label_encoder.fit_transform(df["embarked"])

# Display unique encoded values
print(dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_))))
print(df[["embarked"]].head())
```

```python
# Generate random dates for the "Ticket_Purchase_Date" column
np.random.seed(42)  # For reproducibility
date_range = pd.date_range(start='2020-01-01', end='2022-12-31', freq='D')
df['Ticket_Purchase_Date'] = np.random.choice(date_range, size=len(df))

# Convert "Ticket_Purchase_Date" to datetime format
df['Ticket_Purchase_Date'] = pd.to_datetime(df['Ticket_Purchase_Date'], errors='coerce')

# Extract the year and month from "Ticket_Purchase_Date"
df['Purchase_Year'] = df['Ticket_Purchase_Date'].dt.year
df['Purchase_Month'] = df['Ticket_Purchase_Date'].dt.month

# Display the updated dataframe
print(df[['Ticket_Purchase_Date', 'Purchase_Year', 'Purchase_Month']].head())
```

```python
# Define bins for age categories
bins = [0, 12, 19, 60, float('inf')]
labels = ['Child', 'Teen', 'Adult', 'Senior']

# Create a new "Age_Category" column based on the "Age" column
df['Age_Category'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)

# Display the updated dataframe
df.head()
```
